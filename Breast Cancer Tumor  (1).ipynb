```python
#importing all the required libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
```

```python
#data collection and processing
#(loading the data from sklearn)

from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()

#here,0 represents 'malignant' and 1 represents 'benign'
```

```python
# adding the 'target' column to the dataframe

df = pd.DataFrame(
    np.c_[breast_cancer.data, breast_cancer.target], 
    columns = [list(breast_cancer.feature_names)+ ['target']]
                 )
```

```python
#splitting the data into training data and testing data (x and y)

X = df.iloc[:, 0:-1]
y = df.iloc[:,-1]
```

```python
X.shape, y.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, random_state = 999)
```

```python
X_train.shape, y_train.shape, X_val.shape, y_val.shape
```

```python
#Using Naive Bayes Classifier
clf = GaussianNB()
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_val, y_val)
```
